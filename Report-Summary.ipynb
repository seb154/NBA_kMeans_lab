The dataset that was used was a merged form between NBA dataset and a salaries dataset. Before any modeling, the merged dataset was cleaned and scaled to ensure all features contribute equally. Figure 1 demonstrates the reason behind this decision (x-axis = number of clusters & y-axis = Silhouette Score).
